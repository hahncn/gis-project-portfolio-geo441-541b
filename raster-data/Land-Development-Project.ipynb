## **Visualization**

---

These maps visualize land that was developed between 2010 and 2024 in blue, as well as imperiled species richness in 2024 in pink in Miami, Florida, and Las Vegas, Nevada. Darker shades of pink correlate with higher numbers of endangered species in a given area. Land development, although not visible on a national scale due to the raster resolution, was most often clustered around previously developed areas, i.e. highly populated cities. Miami and Las Vegas were selected for a close-up due to the differing relationships between land development and imperiled species richness around each city.

## **Patterns**

---


Miami has experienced a uniquely large amount of highly concentrated development since 2010, consistent with Florida's rapid population increase. As pictured, blocks of development overlap heavily with areas of high endangered species presence. Las Vegas presents a very different pattern. The greater Las Vegas area saw lots of land development that does not overlap with high numbers of endangered species. There are scattered areas of higher imperiled species richness to the west of Las Vegas, but almost none overlapping the development itself.

## **Findings**

---

Due to these diverging patterns, there does not appear to be a direct relationship between land development and high imperiled species richness based on this data alone. However, there are many factors to consider in regards to biodiversity, such as species richness and biome. Perhaps Florida simply has higher species richness than Nevada and therefore will have a higher number of endangered species. Since this endangered species richness data is presented as raw numbers rather than rates (compared to total number of species present), this relationship is difficult to determine.

Although no straightforward cause-and-effect conclusions can be drawn from these maps, they serve as a reminder that causes for biodiversity loss are often more interwoven and complicated than they first appear.


## **Creation**

---

These maps were created using ArcGIS Pro 3.5 and U.S. land cover raster data from the National Land Cover Database. I performed an Extract by Attributes to produce layers including only pixels classified as "Developed" (Low, Medium, and High Intensity) from the 2024 layer, and only pixels classified as undeveloped (prairie, forest, grassland, etc.) in the 2010 layer.

From there, I used the Raster Calculator to subtract the 2010 raster from the 2024 raster. The result was a raster layer visualizing any land that had been developed (i.e. status changed from "undeveloped" to "developed") between 2010 and 2024.

I overlaid this data with a raster visualizing imperiled species richness from NatureServe to identify patterns and determine if a direct relationship existed between land development and imperiled species richness.
