- **Write a 100-word paragraph which describes the relationship between**
  - a **`Geodatabase`**
  - a **`Feature Dataset`**
  - and **`Feature Classes`**
  
  Refer to **Page 9** of the assignment guide, and other resources online such as:
    - [Fundamentals of the geodatabase](https://pro.arcgis.com/en/pro-app/latest/help/data/geodatabases/overview/fundamentals-of-the-geodatabase.htm)
    - [Feature datasets in ArcGIS Pro](https://pro.arcgis.com/en/pro-app/latest/help/data/feature-datasets/feature-datasets-in-arcgis-pro.htm)

<mark>A Geodatabase is essentially a file storage system used for organizing data that is likely to be used together. Geodatabases can contain Feature Datasets, or collections of related spatial data, such as data layers. Multiple Feature Datasets may exist within the same Geodatabase, and can contain additional tables and network connections. Feature Classes are data layers that live within Feature Datasets, and are used to hold certain clusters of data together. Feature Classes can exist independent of Feature Datasets, although it may be more convoluted that way. To simplify, Feature Classes are individual data layers collected in a Feature Dataset, all stored in a Geodatabase within ArcGIS Pro.</mark>
